```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys
```

```python
sound_file=wave.open('vowel1_1.wav','r')
```

```python
signal=sound_file.readframes(-1)
signal=np.frombuffer(signal,dtype='int16')
fs=sound_file.getframerate()
```

```python
time=np.linspace(0,len(signal)/fs, num=len(signal))
```

```python
plt.figure(1)
plt.plot(time,signal,'k')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.savefig('./Result/Meet_the_Reviewer_Comments/vowel_wave_vs_time.svg',format='svg',dpi=600)
```
