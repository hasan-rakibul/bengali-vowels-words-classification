```python
import librosa
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
Word_PATH="./Word_Data/Word/"
Vowel_PATH="./Vowel_Data/Vowel/"
```

```python
wave, sr= librosa.load(Word_PATH+'Word2/Word2_2.wav', mono=True, sr=None)
wave=wave[::5]
mfcc=librosa.feature.mfcc(wave,sr=sr)

max_len=13
if (max_len > mfcc.shape[1]): #.shape[1], 1-->column
    pad_width=max_len-mfcc.shape[1]
    mfcc=np.pad(mfcc,pad_width=((0,0),(0,pad_width)),mode='constant')
else:
    mfcc=mfcc[:,:max_len]
```

```python
time=np.linspace(0,0+(1/sr)*mfcc.shape[1],num=mfcc.shape[1],endpoint=False) #sampling rate to actual time
plt.plot(np.transpose(mfcc))
#plt.xticks(time.tolist())
plt.xlabel('Time Sample')
plt.ylabel('MFCC')
plt.title('Word')

plt.savefig('./Result/Word2_2_mfcc_vs_time.png',format='png', dpi=1200)
```
