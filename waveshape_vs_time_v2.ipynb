```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys
```

```python
sound_file=wave.open('vowel1_1.wav','r')
```

```python
signal=sound_file.readframes(-1)
signal=np.frombuffer(signal,dtype='int16')
fs=sound_file.getframerate()
```

```python
time=np.linspace(0,len(signal)/fs, num=len(signal))
```

```python
plt.figure()
plt.plot(time,signal,'k')
plt.xlabel('Time (second)')
plt.ylabel('Intensity')
plt.ylim([-6000,8000])
plt.tight_layout()
plt.savefig('./New_without_CNN/vowel_wave_vs_time.pdf',format='pdf',dpi=600)
```
